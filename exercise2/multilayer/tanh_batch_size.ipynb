```python
from mlp import *
```

```python
mlps = {
    MLP(10.0, [10], tanh, 0.01, 1): [[], [], [], []],
    MLP(10.0, [10], tanh, 0.01, 8*8): [[], [], [], []],
    MLP(10.0, [10], tanh, 0.01, 60000): [[], [], [], []],
}
iterations = [100]
```

```python
run_simulation(mlps, iterations);
```

Im mniejszy batch tym częstsze uaktualnianie wag i tym szybsza nauka - wynika to z faktu, że na jedną epokę mamy częstrze uaktualnianie wag. Natomiast przy zbyt dużym batchu (np wielkości całego zbioru jak w przykładzie powyżej) aktualizacja odbywa się raz na epokę co skutkuje większą randomizacją w przebiegu wykresu.
