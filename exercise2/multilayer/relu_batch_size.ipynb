```python
from mlp import *
```

```python
mlps = {
    MLP(1.0, [10], relu, 0.01, 1): [[], [], [], []],
    MLP(1.0, [10], relu, 0.01, 8*8): [[], [], [], []],
    MLP(1.0, [10], relu, 0.01, 60000): [[], [], [], []],
}
iterations = [10] * 10
```

```python
run_simulation(mlps, iterations);
```

Pomimo teoretycznych różnic jakie powinny być zaopserwowane przy zmianie wielkości batcha - im mniejszy tym szybsza nauka, ale jednocześnie bardziej nieregularny przyrost - nie udało się tego zaopserwować na podanym wykresie.
