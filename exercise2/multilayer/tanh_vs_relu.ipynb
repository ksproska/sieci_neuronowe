```python
from mlp import *
```

```python
mlps = {
    MLP(1.0, [10], tanh, 0.01, 64): [[], [], [], []],
    MLP(1.0, [10], relu, 0.01, 64): [[], [], [], []],
}
iterations = [100]
```

```python
run_simulation(mlps, iterations);
```

Model z funkcją aktywacji tanh (tangens hiperboliczny) wolniej się uczy ze względu na zanikający gradient - funkcja ma majwiększy gradient w okolicach 0. W przypadku relu mamy stały gradient (czy też dokładniej 2 gradienty) i jedynym zagrożeniem może być wybuchający gradient przy zbyt dużych wartościach wag.

```python
in_array = np.array(range(-5, 6))/ 2

plt.plot(in_array, tanh(in_array), color = 'red', label="tanh")
plt.plot(in_array, relu(in_array), color = 'blue', label="relu")
plt.title("tanh vs relu")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
