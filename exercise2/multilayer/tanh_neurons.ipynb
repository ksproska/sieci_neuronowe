```python
from mlp import *
```

```python
mlps = {
    MLP(10.0, [10], tanh, 0.01, 64): [[], [], [], []],
    MLP(10.0, [50], tanh, 0.01, 64): [[], [], [], []],
    MLP(10.0, [100], tanh, 0.01, 64): [[], [], [], []],
    MLP(10.0, [500], tanh, 0.01, 64): [[], [], [], []],
}
iterations = [10] * 10
```

```python
run_simulation(mlps, iterations);
```

Linia przerywana oznacza wartości precyzji i błędu dla zbioru treningowego, a linia ciągła dla zbioru testowego.
Jak można zauważyć przy ok 2 iteracjach wartości w zarówno precyzji jak i błędu zaczynają się różnić dla obu zbiorów.
Wynika to z faktu, że model dopoasowuje się do zadanego problemu (zbioru treningowego) i polepsza swoją skuteczność, ale oznacza  to utratę uogólnienia - uczy się "na pamięć". Dla zbioru treninowego poprawanie jest znaczna, przez co sugerowane byłoby w tym miejscu zatrzymać uczenie całkowicie, albo wznowić je z innymi parametrami.
