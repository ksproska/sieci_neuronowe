```python
rows_num = 1
col_num = 3
fig_inches = 6
fig, axs = plt.subplots(rows_num, col_num)
fig.set_size_inches(fig_inches * col_num, fig_inches * rows_num)
axs = np.array(axs).reshape(-1)
x = np.arange(1, 10)
i =1
for ax in axs:
    i+=1
    plt.sca(ax)
    plt.plot(x, x*i + i)

    plt.title(f"Title for {i} plot")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.ylim(0, 35) #zakres osi y analogicznie xlim do x

```
